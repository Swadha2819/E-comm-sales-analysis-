```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import plotly.colors as colors
pio.templates.default = 'plotly_white'
```

```python
data = pd.read_csv("/content/Sample - Superstore.csv", encoding='latin-1')
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year']= data ['Order Date'].dt.year
data['Order Week']= data ['Order Date'].dt.dayofweek
```

```python
data
```

```python
#find monthly sales analysis nd highest sale nd lowest sale in which months
sales_monthly = data.groupby('Order Month')['Sales'].sum().reset_index()

max_row= sales_monthly[sales_monthly['Sales'] == sales_monthly['Sales'].max()]
min_row= sales_monthly[sales_monthly['Sales'] == sales_monthly['Sales'].min()]

fig = px.line(sales_monthly, x='Order Month', y='Sales', title='Monthly Sales Trend',markers=True)

fig.update_layout(xaxis_title='Month', yaxis_title='Sales')

fig.add_scatter(
    x=max_row['Order Month'],
    y=max_row['Sales'],
    mode='markers+text',
    text='Max Sale',
    textposition = 'top center',
    marker=dict(color='green', size=10, symbol= 'circle'),
    showlegend = False
)
fig.add_scatter(
    x=min_row['Order Month'],
    y=min_row['Sales'],
    mode='markers+text',
    marker=dict(color='red', size=10, symbol= 'circle'),
    text='Min Sale',
    textposition = 'bottom center',
    showlegend=False
)
fig.show()
```

```python
sales_category = data.groupby('Category')['Sales'].sum().reset_index()

fig=px.pie(sales_category, values='Sales', names='Category', title='Sales by Category', hole =0.4)

fig.update_layout(legend_title='Category')
fig.update_traces(textposition='inside', textinfo='percent+label',textfont_color='white'
)

fig.show()
```

```python
sales_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()

fig = px.bar(sales_subcategory, x='Sub-Category', y='Sales', title='Sales by Sub-Category')

max_row = sales_subcategory[sales_subcategory['Sales'] == sales_subcategory['Sales'].max()]
min_row = sales_subcategory[sales_subcategory['Sales'] == sales_subcategory['Sales'].min()]

fig.add_scatter(
    x=max_row['Sub-Category'],
    y=max_row['Sales'],
    mode='markers+text',
    text='Max Sale',
    textposition = 'top center',
    marker= dict(color='black',size=15, symbol='circle'),
    showlegend=False
)

fig.add_scatter(
    x=min_row['Sub-Category'],
    y=min_row['Sales'],
    mode='markers+text',
    text='Min Sale',
    textposition = 'bottom center',
    marker= dict(color='black',size=15, symbol='circle'),
    showlegend=False
)

fig.update_layout(xaxis_title='Sub-Category', yaxis_title='Sales')

fig.show()
```

```python
profit_monthly = data.groupby('Order Month')['Profit'].sum().reset_index()

fig = px.line(profit_monthly, x='Order Month', y='Profit', title='Monthly Profit',)

fig.update_layout(xaxis_title='Month', yaxis_title='Profit')

fig.show()
```

```python
profit_category = data.groupby('Category')['Profit'].sum().reset_index()

fig = px.pie(profit_category, values='Profit', names='Category', title='Profit by Category', hole=0.4)

fig.update_layout(legend_title='Category')

fig.update_traces(textposition='inside', textinfo='percent+label', textfont_color='white')

fig.show()
```

```python
profit_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()

fig = px.bar(profit_subcategory, x='Sub-Category', y='Profit', title='Profit by Sub-Category')

fig.update_layout(xaxis_title='Sub-Category', yaxis_title='Profit')

fig.show()
```

```python
profitandsales_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

fig = px.bar(profitandsales_segment, x='Segment', y=['Sales', 'Profit'], barmode='group', title='Profit and Sales by Segment')

fig.update_layout(xaxis_title='Segment', yaxis_title='Amount')

fig.show()
```

```python
profitandsales_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

profitandsales_segment['Sales-Profit Ratio'] = profitandsales_segment['Sales'] / profitandsales_segment['Profit']

print(profitandsales_segment)

print("\n\n",profitandsales_segment[['Segment','Sales-Profit Ratio']])
```
